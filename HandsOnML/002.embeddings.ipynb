## Embedding

- 임베딩(embedding)은 다차원의 데이터를 보다 낮은 차원의 공가능로 표현하는 방법. 자연어 처리에서는 단어, 문장, 문당 등의 텍스트 데이터를 수치화된 벡터 형태로 변환하는 과정을 의미
- One Hot Encoding : 범주형 데이터를 벡터 형태로 변환하는 통계에서 더미변수 처리와 유사. 단어가 많을수록 희속벡터가 됨. 연산 및 유사도 등의 어려움, Encoding의 결과를 One Hot Vector
- 임베팅은 One Hot Encdoing을 연속형 값을 갖게하는 과정
- 임베딩의 용도
  - 검색
  - 군집화
  - 추천
  - 이상 탐지
  - 다양성 측정
  - 분류

```python
import os
from dotenv import load_dotenv

load_dotenv()

from openai import OpenAI

client = OpenAI()
```

### OpenAI의 임베딩 예제

- 적절한 함수를 생성하고, 실행결과로 임베딩 차원수 확인
- text-embedding-3-small 모델 이용

```python
def embedding(input):
    response = client.embeddings.create(
        input = input,
        model = "text-embedding-3-small"
    )

    return response.data[0].embedding

king = embedding("왕")
print( king )
len( king )
```
